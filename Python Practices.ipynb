```python
# EXERCISE
import pandas as pd
student_data = {
    'name':   ['Michael', 'Ada', 'John', 'Sarah', 'David'],
    'subject':   ['Math', 'English', 'Math', 'English', 'Math'],
    'score':   [85, 72, 65, 90, 95]
}
df = pd.DataFrame(student_data)
print('DataFrame:', df)
print('The first 3 rows:', df.head(3))
print('Number of Students who scored above [80]:', df[df['score'] > 80])
grouped_data = df.groupby('subject')
# Written by mik0-logic™
avg_score = grouped_data['score'].mean()
print('The average score for each subject:', avg_score)
below_70 = df[df['score'] < 70]
count = len(below_70)
print(f"Number of Students who scored below [70]:", {count})
```

```python
# EXERCISE
import pandas as pd
employees_data = {
    'name':   ['Micahel', 'Ada', 'John', 'Sarah', 'David'],
    'dept':   ['IT', 'HR', 'Finance', 'IT', 'Finance'],
    'salary':   [5000, 4200, 3900, 6100, 4500]
}
df = pd.DataFrame(employees_data)
print('First 2 rows:', df.head(2))
print('Employees earning more than [$4500]:', df[df['salary'] > 4500])
group_by_dept = df.groupby('dept')
# Written by mik0-logic™
avg_salary = group_by_dept['salary'].mean()
print(f"Average Salary: {avg_salary}")
max_salary = group_by_dept['salary'].max()
print(f"Highest Salary: {max_salary}")
```

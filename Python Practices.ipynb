```python
# Reaction rate calculation
import numpy as np
# Written by mik0-logic™
concentrations = [0.1, 0.25, 0.4, 0.55, 0.7]
# rate law,rates = k.c
c = np.array(concentrations)
k = 0.8
rates = k * c
print('Reaction rates:', rates)
```

```python
# Temperature Conversion
import numpy as np
celsius_temp = [25, 30, 35, 40, 45]
# T(K) = T(℃) + 273.15
C = np.array(celsius_temp)
# Written by mik0-logic™
Kelvin_temp = C + 273.15
print('Celsius Temp:', C)
print('In Kelvin -->', Kelvin_temp)
```

```python
# Heating a Reactor
import numpy as np
T_celsius = np.linspace(25, 100, 10)
# Written by mik0-logic™
kelvin = T_celsius + 273.15
print(f"Celsius: {T_celsius}")
print(f"Kelvin: {kelvin}")
```

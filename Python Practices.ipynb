```python
class Phone:
    def __init__(self, brand, model, starting_batt_percentage):
        self.brand = brand
        self.model = model
        self.starting_batt_percentage = starting_batt_percentage
        print(f"The {self.brand} {self.model} starts with {self.starting_batt_percentage}% battery.")
    def use_phone(self, app_name, drained):
        self.starting_batt_percentage -= drained
        print(f"The app {app_name} drained {drained}% from it")
        print(f"We have just {self.starting_batt_percentage}% left.")

phone_usage = Phone('Apple', 'iphone 15', 100)
phone_usage.use_phone('Instagram', 15)
phone_usage.use_phone('WhatsApp', 10)
```

```python
import pdb

def calculate_interest(principal, rate):
    pdb.set_trace()
    interest = principal * rate
    return interest

result = calculate_interest(10000, 0.05)
print(result)
```

```python
try:
    celsius = float(input('Enter Temperature in Celsius: '))
    to_fahrenheit = (celsius * 9/5)+ 32
    print(f"The celsius Temperature {celsius}°C is same as {to_fahrenheit} in Fahrenheit.")
except ValueError:
    print('Invalid Input!')
    print('Please enter a Valid Number.')
```

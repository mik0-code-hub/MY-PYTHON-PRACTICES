UNDERSTANDING CLASSES AND OBJECTS

```python
# USING THE (BANK SIMULATOR) MINI-PROJECT WE PREVIOUSLY CREATED
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
    def deposit(self, amount):
        self.balance += amount
        print(f"{self.name} deposited ${amount}.\nNew Balance: {self.balance}")
    def withdraw(self, amount):
        self.balance -= amount
        print(f"{self.name} withdrew ${amount}.\nNew Balance: {self.balance}")

alice_acct = BankAccount('Alice', 500)
alice_acct.deposit(100)
alice_acct.withdraw(50)
```

```python
# UNDERSTANDING OOP USING A CAR SIM
class CarSim:
    def __init__(self, year, model, starting_mileage):
        self.year = year
        self.model = model
        self.starting_mileage = starting_mileage
        print(f"The {self.year} {self.model} has started with {self.starting_mileage} miles.")
    def miles_driven(self, miles):
        self.starting_mileage += miles
        print(f"Driven {miles} miles. New Odometer: {self.starting_mileage} miles.")

total_mileage = CarSim(2026, 'Tesla Model 3', 5000)
total_mileage.miles_driven(150)
total_mileage.miles_driven(200)
```
